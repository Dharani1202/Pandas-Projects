```python
import pandas as pd
ds=pd.read_csv('netflix.csv')
ds
```

```python
# if you want to rename the column name then use this!!

ds.rename(columns={'Cast': 'Actors'},inplace=True)

# to know the column name is changed or not ?
ds.columns
```

```python
# If you want to show the column name in the data set then use like this.

ds.to_csv('netflix_updated.csv',index=False)
ds
```

```python
#1. Remove the duplicates if there are any

# to check if any duplicates in the data set

ds[ds.duplicated()]
```

```python
#used to remove the duplicate data

ds.drop_duplicates()
```

```python
#2.	Check whether there are any null values from the dataset

ds.isnull().sum()
```

```python
ds=ds.dropna()
ds
```

```python
ds.isnull().sum()
```

```python
#3.	Print the director name of the show “House of cards”

ds[ds.Title.str.contains('House of Cards')]
```

```python
# it shows the director name of 'house of cards'

ds[ds.Title.str.contains('House of Cards')]['Director'].value_counts()
```

```python
#4.	In which year highest no. of TV shows and movies were released? Show with a bar chart

ds.dtypes
```

```python
ds['Date']=pd.to_datetime(ds['Release_Date'],errors='coerce')
ds
```

```python
# Now we created a new columns as date and year 

ds['Year']=ds['Date'].dt.year
ds
```

```python
ds['Year'].value_counts()
```

```python
ds['Year'].value_counts().plot(kind='bar')
```

```python
#5.	How many TV shows and movies are there in the dataset? Show with count plot

ds.Category.value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(ds.Category)
```

```python
#7.	Show only the titles of all TV shows that were released in India

ds[(ds.Category=='TV Show')&(ds.Country=='India')]
```

```python
# it shows only the title 

ds[(ds.Category=='TV Show')&(ds.Country=='India')]['Title']
```
