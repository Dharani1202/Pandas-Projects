```python
#1.	Show the number of confirmed, death and recovered cases in each region

ds.groupby('State').sum()
```

```python
#2.	Remove all the rows where confirmed cases are less than 10

ds['Confirmed']>=10
```

```python
ds[ds['Confirmed']>=10]
```

```python
# Another method for Greater than 10

ds[~ds['Confirmed']<=10]
```

```python
#3.	In which region maximum number of confirmed cases were recorded

ds['Confirmed'].max()
```

```python
ds[ds['Confirmed']==ds['Confirmed'].max()]
```

```python
#4.	In which region minimum number of confirmed cases were recorded

ds[ds['Confirmed']==ds['Confirmed'].min()]
```

```python
#5.	Region with maximum number of deaths

ds[ds['Deaths']==ds['Deaths'].max()]
```

```python
#6.	Region with minimum number of deaths

ds[ds['Deaths']==ds['Deaths'].min()].head()
```

```python
#7.	Cases in India

ds[ds['Region']=='India']
```

```python
#8.	Sort based on confirmed cases in ascending order

ds.sort_values(by=['Confirmed'],ascending=True)
```

```python
#9.	Sort based on recovered cases in descending order

ds.sort_values(by=['Recovered'],ascending=False)
```

```python
# 10. cases in Germany 

ds[ds['Region']=='Germany']
```

```python
# 11. Top 10 US cases 

ds[ds['Region']=='US'].head(10)
```

```python
#12. Confirmed less than 100 then no of deaths less than 50 and recoverd greater than 50

ds[(ds.Confirmed<=100)&(ds.Deaths<=50)&(ds.Recovered>=50)]
```
